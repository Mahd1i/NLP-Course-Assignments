```python
import nltk
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

words = ['talking', 'talked', 'talks', 'ate', 'eating', 'computed', 'computing']

for word in words:
    print(word, "|", stemmer.stem(word))
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("talking talked talks ate eating computed computing")

for token in doc:
    print(token.text, "|", token.lemma_)
```

```python
text = """Latha is very happy today. She is eating an apple. 
Yesterday she ate a banana and she liked it."""

doc = nlp(text)

# استخراج الـ Lemma لكل كلمة
clean_text = [token.lemma_ for token in doc]

# دمجها مرة أخرى
print(" ".join(clean_text))
```
